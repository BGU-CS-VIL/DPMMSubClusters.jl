```julia
using Plots
using Distributed
using LinearAlgebra
addprocs(2)
@everywhere using DPMMSubClusters
@everywhere using Random
```

## Ploting Function

```julia
function plot_dp_2d(pts, labels)
    plt=Plots.plot()
    Plots.plot!(pts[1,:],pts[2,:], seriestype=:scatter, color = labels, markersize = 3, markerstrokewidth = 0.5)
    return plt
end
```

## Genereate data
10^5 Points, 2D, Generated from a 6 components Gaussian mixture model
with components mean sampled from a normal distribution with isotropic variance scaled by 80.

```julia
Random.seed!(12345)
x,labels,clusters = generate_gaussian_data(10^4,2,6,80.0)
```

## Run Model
With concetration parameter of 100.

```julia
labels,clusters,weights = fit(x,10.0, iters = 100, seed = 1234)
```

## Plot

```julia
plot_dp_2d(x,labels)
```
